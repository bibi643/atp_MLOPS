```python
df['Loser']=df.apply(lambda row:(row['Loser'],row['LRank'],row['elo_loser']),axis=1)
df['Winner']=df.apply(lambda row:(row['Winner'],row['WRank'],row['elo_winner']),axis=1)
df
#Here I created des tuples pour les Winner et Loser. avec le nom, le classement et les points ELO
```

```python
df_game=df[['Winner','Loser']]
#df_game['Players']= df_game.apply(lambda row:(row['Winner'],row['Loser']),axis=1)


df_game['Player1']=pd.DataFrame(
    df_game.values[range(df_game.shape[0]),
                   np.random.randint(
                       0, df_game.shape[1], size=df_game.shape[0])])

df_game

#Randomisation of Player 1
```

```python
df_game['Player2']=''
for i in range(len(df_game)):
    if df_game['Player1'][i]== df_game['Winner'][i]:
        df_game['Player2'][i]=df_game['Loser'][i]
    else:
        df_game['Player2'][i]= df_game['Winner'][i]

#player 2 is added.
```

```python
df_game['WP1']=np.where(df_game['Winner']==df_game['Player1'],1,0)
df_game['WP2']=np.where(df_game['Winner']==df_game['Player2'],1,0)

df_game

#Here we create the column WP1 is 1 if player 1 won. I created WP2 just in case
```

```python
df_game['Player1_name']=''
df_game['Player1_rank']=''
df_game['Player1_ELO']=''
df_game['Player2_name']=''
df_game['Player2_rank']=''
df_game['Player2_ELO']=''



for i in range(len(df_game)):
    df_game['Player1_name'][i]=df_game['Player1'][i][0]
    df_game['Player1_rank'][i]=df_game['Player1'][i][1]
    df_game['Player1_ELO'][i]=df_game['Player1'][i][2]
    df_game['Player2_name'][i]=df_game['Player2'][i][0]
    df_game['Player2_rank'][i]=df_game['Player2'][i][1]
    df_game['Player2_ELO'][i]=df_game['Player2'][i][2]

df_game

#I exploded the tuples so I have columns for player 1 name, player 1 rank and player 1 ELO...same for player2
```

```python
df_game=df_game.drop(labels=['Winner','Loser'],axis=1)
#I delete those columns. If I dont It would be duplicated on the next DF
```

```python
df_concat= pd.concat([df,df_game],axis=1)
df_concat['SetP1']=np.where(df_concat['WP1']==1,df_concat['Wsets'],df_concat['Lsets'])
df_concat['SetP2']=np.where(df_concat['WP2']==1,df_concat['Wsets'],df_concat['Lsets'])

df_concat.info()
#we add sets for each players
```

```python
df_clean=df_concat.dropna()

df_clean
```

```python
#reset the index
df_clean=df_clean.reset_index()
df_clean=df_clean.drop(labels='index',axis=1)
df_clean
```
